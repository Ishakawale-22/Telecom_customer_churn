```python
 # Converted 0 and 1 values od senior citizen to yes/no to make it easier to understand
```

```python
def conv(value):
    if value == 1:
        return 'Yes'
    else:
        return 'No'

df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)       
```

```python
ax = sns.countplot(x = df['Churn'])
ax.bar_label(ax.containers[0])
plt.title('Count of customers by Churn')
plt.show()
```

```python
plt.figure(figsize=(3,4))
gb = df.groupby('Churn').agg({'Churn' : 'count'})
plt.pie(gb['Churn'], labels = gb.index, autopct = '%1.2f%%')
plt.title('Percentage of Churned Customers', fontsize = 10)
plt.show()
```

###### From given piechart we can conclude that 26.54% of our customers have churned out

```python
plt.figure(figsize=(6,4))
sns.countplot(x = df['gender'], data = df, hue = 'Churn')
plt.title('Churn by Gender')
plt.show()
```

```python
counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)
percentages = counts.div(counts.sum(axis=1), axis=0) * 100

#Plot setup
labels = ['No', 'Yes']  # For SeniorCitizen = 0 and 1
churn_labels = percentages.columns.tolist()
colors = ['pink', 'purple']  # Green for No, Red for Yes

fig, ax = plt.subplots(figsize=(4, 4))

# 3. Plot stacked bars
bottom = [0] * len(percentages)
for i, churn_status in enumerate(churn_labels):
    values = percentages[churn_status].values
    ax.bar(labels, values, bottom=bottom, label=churn_status, color=colors[i])
    
    # Add percentage labels
    for j in range(len(values)):
        if values[j] > 0:
            ax.text(j, bottom[j] + values[j]/2, f'{values[j]:.1f}%', 
                    ha='center', va='center', color='white' if values[j] > 20 else 'black', fontsize=10)
    
    # Update bottom for next stack
    bottom = [bottom[j] + values[j] for j in range(len(values))]

# 4. Final touches
ax.set_title('Churn by Senior Citizen (Percentage)')
ax.set_ylabel('Percentage')
ax.set_xlabel('Senior Citizen')
ax.legend(title='Churn')
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```

###### Compartive a greater percentage of people in senior citizen category have churned

```python
import warnings
warnings.filterwarnings("ignore")

# Assuming telecom_cust is your DataFrame with the service columns
services = ['PhoneService', 'MultipleLines', 'InternetService',
            'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
            'TechSupport', 'StreamingTV', 'StreamingMovies']

n = len(services)
cols = 3
rows = (n + cols - 1) // cols

fig, axes = plt.subplots(rows, cols, figsize=(cols*5, rows*4), constrained_layout=True)
axes = axes.flatten()

for ax, col in zip(axes, services):
    sns.countplot(data=df, x=col, ax=ax, palette='Set1', hue = df['Churn'] )
    ax.set_title(col)
    ax.set_xlabel('')
    ax.set_ylabel('Count')
    ax.tick_params(axis='x', rotation=45)

# Turn off any unused axes
for ax in axes[n:]:
    ax.axis('off')

plt.show()
```

##### Fiber optic users show significantly higher churn** than DSL or no-internet groups, while ubscribers to value-added services (like security, backup, tech support, streaming) tend to churn less. Phone service usage has minimal impact on churn, and having multiple lines doesnâ€™t strongly affect churn rates
